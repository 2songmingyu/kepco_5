```python
import pandas as pd
```

```python
# csv 데이터 가져오기
df1 = pd.read_csv('전국 시간별/전국 시간별 전력 사용량(2013-2020).csv',encoding='cp949')
df2 = pd.read_csv('전국 시간별/전국 시간별 전력 사용량(2021).csv',encoding='cp949')
```

```python
# 데이터 프레임 합치기
df = pd.concat([df1,df2],axis=0)
```

```python
# 날짜 데이터에서 년,월,일 추출
df['년']=df['날짜'].str.split('-').str.get(0).astype(int)
df['월']=df['날짜'].str.split('-').str.get(1).astype(int)
df['일']=df['날짜'].str.split('-').str.get(2).astype(int)
```
