```python
import pandas as pd
import numpy as np
```

```python
home = pd.read_csv('주택용_요일특성별_124시_전력소비계수_2008년부터 2021년.csv',encoding='cp949')
```

```python
home.head(3)
```

```python
home['시간대별(1)'][0]='월'
home['시간대별(1)'][1]='요일'
home = home.T
```

```python
home.head(3)
```

```python
# 첫번째 행을 컬럼값으로 변경
home = home.rename(columns=home.iloc[0])
```

```python
# 첫번째 행을 제거
home = home.drop(home.index[0])
```

```python
home.head(3)
```

```python
# map 함수를 이용해 index에서 연도 값만 가져오기
home.index = map(lambda x:x[0],home.index.str.split('.'))
```
