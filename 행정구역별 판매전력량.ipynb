```python
import pandas as pd
```

```python
elec = pd.read_csv('행정구역별_판매전력량_추이_1979년부터 2021년.csv',encoding='cp949').T
elec.head(3)
```

```python
# 첫번째 행을 컬럼값으로
elec = elec.rename(columns=elec.iloc[0])
```

```python
elec.head()
```

```python
# 첫번째 행 제거
elec=elec.drop(elec.index[0])
```

```python
# column 월별(1)을 지역으로 변경
elec = elec.rename(columns={'월별(1)':'지역'})
```

```python
# 개성데이터 삭제
elec = elec[elec['지역']!='개성']
```
